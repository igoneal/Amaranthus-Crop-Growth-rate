```python
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import shap
plt.style.use("ggplot")

import catboost as ctb
```

```python
dataset = pd.read_csv("Amaranthus_growth_rate_A.csv")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,5].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#Here, we are using CatBoostRegressor as a Machine Learning model to fit the data.

model_CBR = ctb.CatBoostRegressor()
model_CBR.fit(X_train, y_train)
print(model_CBR)
```

```python
#Now we have predicted the output by passing X_test and also stored real target in expected_y.

expected_y  = y_test
predicted_y = model_CBR.predict(X_test)
print("Predictions",predicted_y)
```

```python
#Here we have printed r2 score and mean squared log error for the Regressor.

print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))
    
plt.figure(figsize=(5,5))
sns.regplot(expected_y, predicted_y, fit_reg=True, scatter_kws={"s": 100})
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
#X = np.arange(-5, 5, 0.25)
X =expected_y
Y= predicted_y
#Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import shap
from sklearn.metrics import r2_score
from sklearn.metrics import recall_score, accuracy_score, average_precision_score, f1_score, mean_squared_error,mean_absolute_error
plt.style.use("ggplot")

import catboost as ctb
dataset = pd.read_csv("Amaranthus_growth_rate_A.csv")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,5].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
#X, y = make_regression(n_samples=100,n_features=10,n_informative=5,n_targets=1,random_state=42)
#X_train, X_test,\y_train, y_test = train_test_split(X, y,test_size=0.2)
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
y_pred1 = gbr.predict(X_test)
rG = r2_score(y_test,y_pred1)
print("Gradient Boosting - R2: ",rG)

mae = metrics.mean_absolute_error(y_test, y_pred1)
mse = metrics.mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse) # or mse**(0.5)  

print("Mean Absolute Error= %.2f" %mae)
print("Mean square error= %.2f" %mse)
print("Root Mean square Error= %.2f" %rmse)  
```

```python
#Now we have predicted the output by passing X_test and also stored real target in expected_y.
expected_y  = y_test
predicted_y = gbr.predict(X_test)
print("Predictions",predicted_y)
```
